```python
# reading files
f1 = open("C:/Atshan/Atshan_83/ATSHAN_BASE-83_JUL28B_2PH_C.DATA", "r")  
f2 = open("C:/Atshan/Atshan_83/ATSHAN_BASE-83_JUL28B_2PH_D.DATA", "r")  
```

```python
i = 0
  
for line1 in f1:
    i += 1
      
    for line2 in f2:
          
        # matching line1 from both files
        if line1 == line2:  
            # print IDENTICAL if similar
            print("Line ", i, ": IDENTICAL")       
        else:
            print("Line ", i, ":")
            # else print that line from both files
            print("\tFile 1:", line1, end='')
            print("\tFile 2:", line2, end='')
        break
  
# closing files
f1.close()                                       
f2.close()  
```
